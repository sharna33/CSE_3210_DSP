# ***Topic: Wezney Method***

# **Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Input Signals**

```python
# taking input signals x(n) & h(n)

x_n = np.arange(1, 11, 1)
x_origin = 2
h_n = np.arange(1, 11, 1)
h_origin = 3

print(x_n)
print(h_n)
```

# **Product of Matrices**

```python
matrix = np.zeros((len(x_n), len(h_n)))

for i in range(len(x_n)):
  for j in range(len(h_n)):
      matrix[i][j] = h_n[i] * x_n[j]

print(matrix)
```

# **Sum of the diagonals**

```python
# Sum the diagonal elements of the matrix to compute y(n)
y_n = []

# Sum diagonals starting from the top row and left column

for k in range(len(matrix)):
    diagonal = 0
    for i in range(k + 1):
        diagonal += matrix[i, k - i]
    y_n.append(diagonal)

# Sum diagonals starting from the right column (excluding the main diagonal)

for k in range(1, len(matrix)):
    diagonal = 0
    for i in range(k, 4):
        diagonal += matrix[i, 4 - (i - k) - 1]
    y_n.append(diagonal)

print("Convoluted Signal : ", y_n)

result_origin = x_origin + h_origin
print("Origin of the covoluted signal = ", result_origin)
```

# **Plotting the Convoluted Signal**

```python
# plot
fig, axs = plt.subplots(3, 1, figsize=(10, 8))
plt.suptitle("Convolution Wezney Method", color='black')

# Apply styles
plt.style.use('ggplot')
plt.rcParams['grid.color'] = 'grey'

# Plot the input signal
axs[0].stem(range(len(x_n)), x_n, linefmt='b-', markerfmt='co', basefmt='c-')
axs[0].plot(x_origin, x_n[x_origin], 'ro')
axs[0].set_title("Input Signal X", color='black')
axs[0].legend(["Origin of x_n", "x[n]"], labelcolor='black')
axs[0].grid(True)

# Plot the impulse response
axs[1].stem(range(len(h_n)), h_n, linefmt='b-', markerfmt='co', basefmt='c-')
axs[1].plot(h_origin, h_n[h_origin], 'ro')
axs[1].set_title("Impulse Response h_n", color='black')
axs[1].legend(["Origin of h_n", "h[n]"], labelcolor='black')
axs[1].grid(True)

# Plot the output signal
axs[2].stem(range(len(y_n)), y_n, linefmt='b-', markerfmt='co', basefmt='c-')
axs[2].plot(result_origin, y_n[result_origin], 'ro')
axs[2].set_title("Convulated Signal y_n", color='black')
axs[2].legend(["Origin of y_n", "y[n]"], labelcolor='black')
axs[2].grid(True)

plt.show()
```
